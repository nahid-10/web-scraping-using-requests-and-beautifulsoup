```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
```

```python
url="https://en.wikipedia.org/wiki/Demographics_of_Bangladesh"
```

```python
response=requests.get(url,headers={'user-agent':"Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/138.0.0.0 Safari/537.36"})
```
